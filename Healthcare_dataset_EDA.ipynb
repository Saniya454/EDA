```python
# distribution analysis of medical conditions using histogram
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
ax = sns.histplot(df["Medical Condition"], bins=30, binwidth=200,alpha=0.7)
ax.set_ylim(9000, 10000)
plt.show()
```

```python
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution of Patients")
plt.show()
```

```python
ax=sns.countplot(y=df["Medical Condition"], order=df["Medical Condition"].value_counts().index)
ax.set_xlim(9000, 10000)
plt.title("Most Common Medical Conditions")
plt.show()
```

```python
# yearly admission trend
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"], format="%d-%m-%Y")
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"], format="%d-%m-%Y")
df["Year"] = df["Date of Admission"].dt.year
df.groupby("Year")["Age"].count().plot(kind="line", marker="o")
df["Year"] = df["Date of Admission"].dt.year
plt.title("Yearly Admissions Trend")
plt.show()
```

```python
top_hospitals = df["Hospital"].value_counts().nlargest(10)  # Top 10 hospitals

plt.figure(figsize=(10,5))
sns.barplot(x=top_hospitals.index, y=top_hospitals.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Top 10 Hospitals by Admissions")
plt.xlabel("Hospital")
plt.ylabel("Number of Admissions")
plt.show()
```

```python
# admission pattern
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Admission Type", palette="coolwarm",legend="False")
ax.set_ylim(17000, 20000)
plt.title("Distribution of Admission Types")
plt.xlabel("Admission Type")
plt.ylabel("Number of Admissions")
plt.show()
```

Hypothesis 1: Emergency admissions have higher billing amounts than elective admissions

```python
plt.figure(figsize=(8,5))
ax=sns.boxplot(data=df, x="Admission Type", y="Billing Amount", palette="coolwarm")
ax.set_ylim(10000, 40000)
plt.title("Billing Amount by Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Billing Amount")
plt.show()
```

```python
from scipy.stats import ttest_ind

emergency = df[df["Admission Type"] == "Emergency"]["Billing Amount"]
elective = df[df["Admission Type"] == "Elective"]["Billing Amount"]

t_stat, p_value = ttest_ind(emergency, elective, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

No. We can see that billing amount has no dependency on admission type

 Hypothesis: Older patients are most likely admitteld due to chronic conditions: Cancer, Obesity, Diabetes

```python
df["Age Group"] = df["Age"].apply(lambda x: "Senior (60+)" if x >= 60 else ("Adult (40-59)" if 40 <= x <= 59 else "Young (18-39"))

plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Medical Condition",hue="Age Group", palette="coolwarm")
ax.set_ylim(0,4000)
plt.title("Medical Conditions by Age Group")
plt.xlabel("Medical Condition")
plt.ylabel("Count")
plt.show()
```

There is no trend we can see that older patients have most likely to be admitted for chronic conditions
